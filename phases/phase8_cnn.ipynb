```python
# --- Reflection: When Should You Use a CNN? ---
# Write markdown observations (or print statements) addressing:
#
#   1. CNNs assume spatial locality: nearby pixels are more related than distant ones.
#      This holds for images (adjacent pixels share color/texture) but NOT for tabular features
#      (Temperature and Distance are not spatially related just because we placed them next to each other).
#
#   2. With only 29 samples, the model is more likely to memorize than generalize.
#      CNNs typically need thousands of images to learn useful filters.
#
#   3. The real value of this exercise: you now know how to construct a CNN,
#      interpret its architecture, and reason about when convolution is appropriate.
#      Phase SP2 in the Pokémon superphase applies these ideas to actual image/sequence data.
```
